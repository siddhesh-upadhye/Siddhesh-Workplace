# MatplotLib

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
#scatterplot

plt.scatter(x,y,c='b')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Scatterplot')
```

```python
y=x*x
```

```python
plt.plot(x,y,'b*',linestyle='dashed')
```

```python
#subploting
plt.subplot(1,2,1)
plt.plot(x,y,'r*')
plt.subplot(1,2,2)
plt.plot(x,y,'bo')
```

```python
np.pi
```

```python
#sin Wave
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
```

```python
#bar_plot
a=[1,3,5]
b=[2,5,4]
a1=[2,4,6]
b1=[1,7,8]
plt.bar(a,b)
plt.bar(a1,b1)
```

```python
#pie Plot
labels='C++','Java','Python','HTML','.Net'
sizes=[130,175,154,182,220]
colors=['gold','Green','Red','Pink','Blue']
explode=(0.15,0,0,0,0)
plt.pie(sizes,labels=labels,colors=colors,explode=explode,autopct='%1.1f%%',shadow=True)
```

```python
df1=sns.load_dataset('iris')
df1
```

```python
#Univariate analysis
df_setosa=df1.loc[df1['species']=='setosa']
df_virginica=df1.loc[df1['species']=='virginica']
df_versicolor=df1.loc[df1['species']=='versicolor']
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']))
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']))
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']))
```

```python
#Bivariate analysis
sns.FacetGrid(df1,hue='species',size=5).map(plt.scatter,'petal_length','sepal_width').add_legend()
```

```python
#Multivariate Analysis
sns.pairplot(df1,hue='species',size=2)
```

```python
sns.pairplot(df1,size=2)
```
